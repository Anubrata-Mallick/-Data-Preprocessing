# Imbalanced Data Set

A data set with an un equal class distribution

```python
import pandas as pd
import numpy as np
```

```python
credit_card_data = pd.read_csv("/content/credit_data.csv")
```

```python
credit_card_data.head()
```

```python
# distributionof 2 classes
credit_card_data['Class'].value_counts()
```

**This is a Highly Imbalanced Data**


0 --> Legit Transaction

1 --> Fraudulent Transaction

```python
# separating the legit and fraudulent transactions
legit = credit_card_data[credit_card_data.Class == 0.0]
fraud = credit_card_data[credit_card_data.Class == 1.0]
```

```python
print(legit.shape)
print(fraud.shape)
```

Undr-sampling

Build a sample dataset containing similar distribution of Legit & Fraudulent Transactions


Number of fraud = 177

```python
legit_sample = legit.sample(n=177)
print(legit_sample.shape)
```

Concatinate 2 Data Frame

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```
